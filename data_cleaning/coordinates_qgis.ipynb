### coordinates for qgis

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
qg_co_df = pd.read_excel("../coordinates/coordinaten_vistelius.xlsx", index_col=0)
```

```python
qg_co_df
```

```python
qg_co_df['Lat_sec'] = qg_co_df['Lat_sec'].fillna(0)
qg_co_df['Long_sec'] = qg_co_df['Long_sec'].fillna(0)
qg_co_df['past_mer'] = qg_co_df['past_mer'].fillna(0)
```

```python
# Could also have used the '.as_type(str)' method on each column 
# to convert them to string format but 'map(str)' also does the job
co_Lat_df = qg_co_df["Lat_deg"].map(str) + "°" + qg_co_df["Lat_min"].map(str) + "'" + qg_co_df["Lat_sec"].map(str) + '"' 
co_Long_df = qg_co_df["Long_deg"].map(str) + "°" + qg_co_df["Long_min"].map(str) + "'" + qg_co_df["Long_sec"].map(str) + '"' 
co_merididan = qg_co_df["past_mer"].map(str)
```

```python
co_merididan
```

```python
co_final = pd.concat([co_Lat_df, co_Long_df, co_merididan], axis=1, sort=False)
```

```python
co_final
```

```python
co_final = co_final.rename({0: "Latitude", 
                            1: "Longitude"}, 
                           axis=1)
# co_final = co_final.rename({1 : "Longitude"}, axis=1)
```
