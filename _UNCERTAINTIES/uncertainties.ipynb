```python
x = uncert_df.loc[(uncert_df['sum'] >= 99.75) & (uncert_df['sum'] < 100.50)]
y1 = uncert_df.loc[(uncert_df['sum'] >= 100.75)]
y2 = uncert_df.loc[(uncert_df['sum'] < 99.50)]
z1 = uncert_df.loc[(uncert_df['sum'] >= 99.5) & (uncert_df['sum'] <99.75)]
z2 = uncert_df.loc[(uncert_df['sum'] >= 100.5) & (uncert_df['sum'] < 100.75)]
```
